```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1=pd.read_csv("table/CNN_Fashion_MNIST_As.csv",index_col=0)
df2=pd.read_csv("table/CNN_Fashion_MNIST_Cr.csv",index_col=0)
df3=pd.read_csv("table/Resnet50_Fashion_MNIST_As.csv",index_col=0)
df4=pd.read_csv("table/Resnet50_Fashion_MNIST_Cr.csv",index_col=0)
df5=pd.read_csv("table/Resnet50_Cifar10_As.csv",index_col=0)
df6=pd.read_csv("table/Resnet50_Cifar10_Cr.csv",index_col=0)
df7=pd.read_csv("table/Renet50_Cifar100_As.csv",index_col=0)
df8=pd.read_csv("table/Renet50_Cifar100_Cr.csv",index_col=0)
```

```python
df1_new=df1
df2_new=df2
df1_new.loc['ClassWisePredictor','Margin']=None
df2_new.loc['ClassWisePredictor','Margin']=None
# df1_new= df1_new.sub(1, axis=0)
df2_new= df2_new.sub(0.94, axis=0)
fig, axs = plt.subplots(1,2, figsize=(12, 4))

df1_new.plot(kind='bar', ax=axs[0],rot=0)
axs[0].set_title('Average Size by Predictor and Score Function')
axs[0].set_xlabel('Predictor')
axs[0].set_ylabel('Average Size')
axs[0].legend(title="Score Function")

df2_new.plot(kind='bar', ax=axs[1],rot=0)
axs[1].set_title('Coverage Rate by Predictor and Score Function')
axs[1].set_xlabel('Predictor')
axs[1].set_ylabel('Coverage Rate')
axs[1].legend(title="Score Function")

plt.tight_layout()
plt.show()
```

```python
sns.set(style="whitegrid")

data = {
    'Dataset': ['Fashion-MNIST', 'Cifar10', 'Cifar100', 'Fashion-MNIST', 'Cifar10', 'Cifar100'],
    'Metric Value': [1.1594, 1.7200, 19.8032/10] + [0.9510, 0.9496, 0.9530],
    'Metric': ['Average Size']*3 + ['Coverage Rate']*3
}

df = pd.DataFrame(data)

g = sns.FacetGrid(df, col="Metric", sharey=False, height=4, aspect=1.5)
g.map(sns.barplot, "Dataset", "Metric Value", palette='muted')

g.fig.suptitle('SplitPredictor using THR Function Metric Comparisons', fontsize=16, fontweight='bold')

plt.subplots_adjust(top=0.85)

plt.show()
```
